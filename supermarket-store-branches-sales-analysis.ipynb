```python
np.random.seed(42)
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(data, test_size=0.2, random_state=42)
```

```python
corr_matrix = data.corr()
corr_matrix["Store_Sales"].sort_values(ascending=False)
```

```python
from pandas.plotting import scatter_matrix

attributes = ["Store_Area"  ,"Items_Available"  ,"Daily_Customer_Count",  "Store_Sales"]
scatter_matrix(data[attributes], figsize=(12, 8))
save_fig("scatter_matrix_plot")
```

```python
datain = train_set.drop("Store_Sales", axis=1) # drop labels for training set
#datain = datain.drop("Store ID", axis=1) # drop labels for training set

result = train_set["Store_Sales"].copy()
```

```python
datain.head()


```

```python
result.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
        ('std_scaler', StandardScaler()),
    ])

datain_tr = num_pipeline.fit_transform(datain)
```

```python
datain_tr
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(datain_tr, result)
```

```python
some_data = datain.iloc[:5]
some_labels = result.iloc[:5]
some_data_prepared = num_pipeline.fit_transform(some_data)
print("Predictions:", lin_reg.predict(some_data_prepared))
```

```python
print("Labels:", list(some_labels))
```

```python
from sklearn.metrics import mean_squared_error

predictions = lin_reg.predict(datain_tr)
lin_mse = mean_squared_error(result, predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.metrics import mean_absolute_error
predictions = lin_reg.predict(datain_tr)
lin_mae = mean_absolute_error(result,predictions)
lin_mae
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(datain_tr, result)
```

```python
predictions = tree_reg.predict(datain_tr)
tree_mse = mean_squared_error(result, predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```
